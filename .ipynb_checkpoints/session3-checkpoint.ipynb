##### Lock-in Amplifier Setup:

1. Turn on lock-in amplifier
2. Set INT/EXT/2f to EXT
3. Ensure photodiode output plugged into Reference Input
4. Plug output of Lock-in into digital multimeter (set to 20V DC)
5. Plug PMT output into lock-in input
6. Set time constant dial to 400ms
7. Set phase switches all to 0
8. Select 100-1k frequency range
9. Scan using the monochrometer until a clean modulated signal is found
10. Turn on 90 degree switch
11. Adjust the phase so DMM reads 0V
12. Turn the 90 degree phase switch back to 0
13. Find the largest possible signal from DMM by adjusting the monochrometer, keep an eye out for the overload light
14. Adjust the sensitivity such that the largest signal doesn't overload the lock-in

#### Uncertainty Analysis:

##### Monochrometer Uncertainty:

The linear stage has an uncertainty of $ \pm 0.0001 in$, very small compared to our readings so we will disregard it.  The largest source of uncertainty when converting the monochrometer reading in inches to wavelength in nanometers is the goodness of our calibration.  

From session 2 we discovered that the relationship between the monochrometer reading in inches to the allowed wavelength in nanometers was:  
$$X_{nanometers} = m_s \times X_{inches} - 62.20915$$ 
$$ m_s = 1002 \pm 76 $$

Uncertainty is as follows:
$$ y = mx + b$$
$$\delta y = y \sqrt{ \frac{\delta y}{\delta m}^2 \delta m ^2} = x y \delta m$$

This resulted in an average uncertainty of about 50 nanometers.  As such it's difficult to correlate the measured intensities to emission bands of neon and helium.  

#### Observations:  

We checked whether the lower signal intensity when the cavity is lasing corresponded to a positive or negative output from the DMM.  To do this, we found a nicely modulated signal using the monochrometer and the oscilloscope, and placed a piece of paper infront of the outgoing beam to stop laser.  

**When blocked, the signal becomes less negative and gave us a positive voltage from the DMM and lock-in amplifier.**

Since the stronger the signal the more negative it is, when blocked the signal becomes less negative (weaker), the positive signal from the lock-in indicates that the emission get stronger when unblocked!  Negative signal from lock-in indicates that the emission gets weaker when cavity is unblocked.  

From session 2, we saw that some emission bands are much larger than others.  To compare how much the modulated signal changes we need to normalize the amplitude of modulated signal to the strength of the steady state signal.  

$$ A_{normalized} = \frac{A_{lock-in}}{V_{steady state}}$$
