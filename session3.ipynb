## Population Inversion:

If the laser beam must make a round trip between the two cavity mirrors 50 times before an equilibrium population inversion is produced, then the time it takes for a stable population inversion is as follows:

$$ 1 cycle = 2d$$
$$ d = 42 cm $$
$$ c = 3\times10^8 m/s$$
$$ 50 \times 1 cycle = 100 \times d = 42 m$$
$$ t = \frac{42 m}{3\times10^8 m/s} = 1.4\times10^-7 s$$

Comparing this time to population equilibrium, to the frequency of the chopper and how much time the cavity has blocked and unblocked:

$$f = 100 Hz$$
$$ T = 1/f = 0.01s$$

So the chopper produces an unblocked beam for half the period, which is $0.005s$.  From a comparison between the time required to create population equilibrium, to the time that the cavity is uncovered by the chopper, we can see that the population inversion will be satisfied for almost all of the cavity being uncovered.  Only the first $1.4 \times 10^-7 s$ will be a transient population, and everything after that is steady states!
